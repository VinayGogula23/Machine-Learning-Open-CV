```python
c=0
for i in cats:
    c=c+1
    image=cv2.imread(i,0)
    cv2.imwrite(r"C:\Users\Gogula Vinay\Desktop\Innomatics labs\Data Science\Machine Learning\Animal Assignment\cat\{}.jpg".format(c),image)
```

```python
# Converting cat images into Vectors
```

```python
url=r"C:\Users\Gogula Vinay\Desktop\Innomatics labs\Data Science\Machine Learning\Animal Assignment\cat"
x1=[]
for i in range(1,12500):
    x1.append(url+'\\'+str(i)+'.jpg')
```

```python
l1=[]
for i in x1:
    image=cv2.imread(i)
    image=cv2.resize(image,(100,100))
    l1.append(image.flatten().tolist())
```

```python
catdf=pd.DataFrame(l1)
catdf['label']='cat'
catdf
```

```python
# Converting dog images into Vectors
```

```python
url=r"C:\Users\Gogula Vinay\Desktop\Innomatics labs\Data Science\Machine Learning\Animal Assignment\dog\dog"
y1=[]
for i in range(1,12500):
    y1.append(url+str(i)+'.jpg')
```

```python
l1=[]
for i in y1:
    image=cv2.imread(i)
    image=cv2.resize(image,(100,100))
    l1.append(image.flatten().tolist())
```

```python
dogdf=pd.DataFrame(l1)
dogdf['label']='dog'
dogdf
```

```python
finaldf=pd.concat([catdf,dogdf])
```
