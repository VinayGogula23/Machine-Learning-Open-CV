```python
import cv2
import numpy as np
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

```python
v2=cv2.VideoCapture(0)
count=0
while True:
    count=count+1
    su,img6=v2.read()
    print(su)
    cv2.imshow("video",img6)
    cv2.imwrite(r'C:\Users\Gogula Vinay\Desktop\Innomatics labs\Data Science\Machine Learning\ML day 5 assignment\Fives\{}.jpg'.format(count),img6)
    if cv2.waitKey(1)==ord("a"):
        break
cv2.destroyAllWindows()
```

```python
# Using the ave code we are going to conver the video in images.
```

```python
url=r"C:\Users\Gogula Vinay\Desktop\Innomatics labs\Data Science\Machine Learning\ML day 5 assignment\Ones"
x1=[]
for i in range(1,230):
    x1.append(url+'\\'+str(i)+'.jpg')
```

```python
l1=[]
for i in x1:
    image=cv2.imread(i,0)
    image=cv2.resize(image,(100,100))
    l1.append(image.flatten().tolist())
```

```python
oneimg=pd.DataFrame(l1)
oneimg['label']='one'
oneimg
```
